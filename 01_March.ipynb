```python
# Are a few products driving most of the revenue?
# total revenue per product
product_sales = df.groupby('product_name')['sales'].sum().sort_values(ascending = False)
# cumulative revenue
cumulative_sales = product_sales.cumsum()
total_sales = product_sales.sum()
# percentage of total revenue
cumulative_percentage = cumulative_sales/total_sales*100
cumulative_percentage
top_80_products = cumulative_percentage[cumulative_percentage <= 80]

print("Number of products contributing to ~80% revenue:", len(top_80_products))
print("Total products in dataset:", len(product_sales))
print("percentage:",(len(top_80_products)/len(product_sales))*100)
```
