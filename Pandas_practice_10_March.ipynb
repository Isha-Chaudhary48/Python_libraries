```python
def val_counts(CustomerId_Count):
    if CustomerId_Count> 30:
        return "Great"
    elif CustomerId_Count > 15:
        return "Average"
    elif CustomerId_Count > 5:
        return 'Below'
    else:
        return 'New Customer'
```

```python
customer_freq = df['Customer ID'].value_counts()
df['Cust_Category']=df['Customer ID'].map(customer_freq).apply(val_counts)
```

```python
def city_cat(city):
    if city > 500:
        return "Excellent"
    elif city > 300:
        return "Good"
    elif city > 100:
        return "Average"
    else:
        return "Below Average"
        
```
