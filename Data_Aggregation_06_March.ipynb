```python
def check():
    group = input("Enter column of the group")
    column = input("Enter the column of agreegation")
    choice = int(input("Enter choice 1. count, 2. sum , 3. mode"))
    if choice == 1:
        print(df.groupby(group)[column].count())
    elif choice ==2:
        print(df.groupby(group)[column].sum())
    elif choice == 3:
        print(df.groupby(group)[column].agg(lambda x : x.mode()[0]))
    else:
        print("Invalid choice")
check()

    
```
