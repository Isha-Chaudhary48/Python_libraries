```python
import matplotlib.pyplot as plt
```

```python
df.info()
```

```python
# 1. Insight
# which state produces the most honey
maxHoney=df.groupby('state')['production'].sum().sort_values(ascending = False).nlargest(10)
plt.figure(figsize=(15,6))
plt.bar(maxHoney.index,maxHoney.values)
plt.xlabel("State")
plt.ylabel("Total honey Production")
plt.show()
# max honey production in these states means strong beekeeping industries and favourable environment
```

```python
# 2. Insight
# Honey Production Trend Over Years
Honey_Production=df.groupby('year')['production'].sum()
plt.plot(Honey_Production.index,Honey_Production.values)
plt.xlabel("Years")
plt.ylabel("Production")
plt.show()
# Honey production was relatively high at 195 and then gradually decreases in following years
# it can be due to environment changes ,reduction in bee colonies,increased pesticide use may have negatively affected bee populations and honey production
# changes in agricultural practices 


```

```python
# 3. Insight
# Average Honey Price Trend Over Years
average_honey_price=df.groupby('year')['average_price'].mean()
plt.plot(average_honey_price.index,average_honey_price.values)
plt.xlabel("Year")
plt.ylabel("Average Price")
```

```python
# States With Highest Number of Bee Colonies
BeeColonies=df.groupby('state')['colonies_number'].sum()
plt.figure(figsize=(25,7))
plt.bar(BeeColonies.index,BeeColonies.values)
# Certain states maintain significantly more bee colonies, indicating strong beekeeping activity.
```

```python
# Relationship Between Colonies and Production
plt.scatter(df['colonies_number'],df['production'])
plt.xlabel("Colonies")
plt.ylabel("Production")
plt.show()
```

```python
# Highest Yield Per Colony
df.groupby('state')['yield_per_colony'].mean().sort_values(ascending=False)
```

```python
# Total Value of Honey Production by State
df.groupby('state')['value_of_production'].sum()
```

```python
# Production vs Price Relationship
plt.scatter(df['production'],df['average_price'])
plt.xlabel("Production")
plt.ylabel("Average Price")
plt.show()
```

```python
# Stock Levels Over Time
stock_level=df.groupby('year')['stocks'].sum()
plt.bar(stock_level.index,stock_level.values)
plt.xlabel("Year")
plt.ylabel("Stocks")
plt.show()
```

```python
# Production Distribution
plt.hist(df['production'])
plt.show()
```

```python
# Correlation Between Variables
df['yield_per_colony'].corr(df['production'])
```

```python
# Year With Highest Honey Production
df.groupby('year')['production'].sum().idxmax()
```

```python
# State With Highest Economic Value
df.groupby('state')['value_of_production'].sum().sort_values(ascending=False)
```

```python
# Multiplots
fig,ax=plt.subplots(1,3,figsize=(15,6))
ax[0].scatter(df['production'],df['stocks'])
ax[1].bar(state_avg_price.index,state_avg_price.values)
ax[2].hist(df['value_of_production'])
plt.show()

```
