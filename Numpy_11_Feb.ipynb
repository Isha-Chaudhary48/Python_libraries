```python
expenses = np.array(
    [
        [130,240,200,180,320], # food
        [100,200,49,80,90]  , # travel
        [1000,300,320,800,900] # shopping
        
        
    ]
)

```

```python
print("Total expense",np.sum(expenses))
print("Category wise expense",np.sum(expenses,axis = 1))
print("day wise total",np.sum(expenses,axis = 0))
print("highest expense day",np.max(np.sum(expenses,axis = 0)))
```
