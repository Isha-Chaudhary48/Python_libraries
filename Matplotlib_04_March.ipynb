```python
df.drop(index=12,inplace=True)
```

```python
df
```

```python
df["Dept"]=['HR','IT','FINANCE','HR','IT','SALES','FINANCE','IT','IT','HR','SALES','HR']
```

```python
df
```

```python
# Pie chart
count_Dept=df['Dept'].value_counts()
```

```python
plt.pie(count_Dept,labels = count_Dept.index,autopct="%1.2f",explode=[0,0.1,0,0.1])
plt.axis("equal")
plt.show()
```

```python
df["Age"]=[18, 22, 25, 30, 19, 35, 28, 24, 40, 21, 27, 32]
```

```python
df
```

```python
# Scatter plot : help to analyze direct relationship
# A scatter plot is used to show the relationship between two numerical variables.
plt.scatter(df["Age"],df["Salary"],color="blue")
plt.show()
```

```python
sort_age = df["Age"].sort_values()
```

```python
# Line plot
# when we used the direct dataframe of age the graph is randomaized if we sort the age then we look up proper graph but not recommended 
plt.plot(sort_age,df["Salary"],color='orange',marker="o",linewidth=2)
plt.grid()
plt.show()
```

```python
# Bar chart
plt.bar(df['Age'],df['Salary'],color='orange',align="edge")
```

```python
# Bivariate: numerical : categorical columns
```

```python
# salary distributed according to the department
Fin_Salary=df[df['Dept']=='FINANCE']['Salary']
Hr_Salary = df[df['Dept']=='HR']['Salary']
It_Salary = df[df['Dept']=='IT']['Salary']
Sales_Salary = df[df['Dept']=='SALES']['Salary']
Fin_Salary
```

```python
# Box plot
plt.boxplot([Fin_Salary,Hr_Salary,It_Salary,Sales_Salary],tick_labels=['FINANCE','HR','IT','SALES'])
plt.show()
```

```python
# pie Chart
salary_by_dept = df.groupby("Dept")["Salary"].sum()
salary_by_dept
```

```python
plt.pie(salary_by_dept,labels=salary_by_dept.index,autopct="%1.2f",shadow=True,explode=[0,0.1,0,0])
plt.show()
```

```python
# bar plot
hr_mean = Hr_Salary.mean()
fin_mean = Fin_Salary.mean()
sal_mean= Sales_Salary.mean()
it_mean = It_Salary.mean()
plt.bar(["HR","FINANCE","SALES","IT"],[hr_mean,fin_mean,sal_mean,it_mean],color=["green","blue","yellow","red"])
plt.grid()
plt.show()
```

```python
# Multivarient Analysis : 3 numerical columns
```

```python
df
```

```python
df["Experience"]=[1, 2, 3, 5, 4, 6, 8, 10, 7, 12, 9, 15]
```

```python
df
```

```python
# Bubble plot:
# when we have 3 numerical columns
plt.scatter(df["Age"],df["Salary"],s=df["Experience"]*20,marker="o",color="skyblue",edgecolors="black")
plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.grid()
plt.show()
```

```python
# when we have 2 numerical and 1 categorical column
```

```python
plt.scatter(df['Age'],df['Salary'],c=df['Dept'].map({'HR':'yellow','FINANCE':'red','SALES':'blue','IT':'green'}))
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age vs Salary vs Dept')
plt.show()
```

```python
color ={'HR':'yellow','FINANCE':'red','SALES':'blue','IT':'green'}
for dept,col in color.items():
    df_dept = df[df["Dept"]==dept]
    plt.scatter(df_dept['Age'],df_dept["Salary"], c = col , label = dept)
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary vs Department")
plt.legend(title="Department")
plt.show()
```

```python
# object oriented API
```

```python
fig,axs =plt.subplots(1,3,figsize=(14,5))
# line plot
axs[0].plot(sort_age,df["Salary"],color="red",marker="o",linewidth="2",markersize = 2)
axs[0].grid()
axs[0].set_title("Line plot")
axs[0].set_xlabel("Age")
axs[0].set_ylabel('Salary')
# histogram
axs[1].hist(df["Salary"],bins = 5,color = "skyblue")
axs[1].set_title("Histogram")
axs[1].set_xlabel("Salary")
axs[1].set_ylabel("Frequency")
# box plot 
axs[2].boxplot(df["Salary"])
axs[2].set_title("Box plot")


# inorder to save the image to use in the presentation or somewhere else
plt.savefig("multipleplots.png")
plt.show()

```

```python
# to show all numeric columns in one chart
data = {
    'Year':[2020,2021,2022,2023],
    'Sales':[100,150,200,250],
    'Profit':[20,30,40,50],
    'Expenses':[80,120,160,200]
}
```

```python
df2 = pd.DataFrame(data)
df2
```

```python
plt.plot(df2["Year"],df2["Sales"],label="Sales")
plt.plot(df2["Year"],df2["Profit"],label = "Profit")
plt.plot(df2["Year"],df2["Expenses"],label = "Expenses")
plt.title("Financial Analysis")
plt.xlabel("Year")
plt.ylabel("Amount")
plt.legend()
plt.show()
```
