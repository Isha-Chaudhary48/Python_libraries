```python
name = ['amit','sumit','kunal']
age = [23,22,25]
subjects = ['english','science','maths','python']
```

```python
marks=[
    [56,67,56,78],
    [34,56,78,90],
    [45,32,67,87]
]
```

```python
df_zip3 = pd.DataFrame(
    [(name,age,*marks)
     for name,age,marks in zip (name ,age,marks)]
    ,columns=['name','age']+subjects
)
```

```python
df_zip3
```
