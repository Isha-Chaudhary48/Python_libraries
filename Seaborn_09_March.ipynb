```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
penguins['species'].value_counts()
```

```python
penguins['island'].value_counts()
```

```python
sns.scatterplot(data = penguins,x='flipper_length_mm',y='body_mass_g',hue='sex')
```

```python
# set_style : background style
#sns.set_style('ticks')
sns.set_style('whitegrid') 
#sns.set_style('dark')
#sns.set_style('darkgrid')
sns.scatterplot(data = penguins,x='flipper_length_mm',y='body_mass_g',hue='island')
# set_context used to control the scaling of the plot elemtes such as text size , line thickness, marker size, tick size
# paper	Small plots for research papers
#notebook	Default size for Jupyter notebooks
#talk	Medium size for presentations
#poster	Very large plots for posters
sns.set_context("talk")
sns.despine(top = True,bottom=True)
# despine removes some of the border lines to make the graph look cleaner and professinal
sns.set_palette("Set1")
# for color theme of plots
# deep,muted,bright,pastel,dark,colorblind
```

```python
sns.scatterplot(data = penguins,x='flipper_length_mm',y='body_mass_g',hue='island',palette='Dark2')
sns.set_context("paper")
```

```python
sns.scatterplot(data = penguins,x = 'species',y='body_mass_g',hue='island',style='sex',alpha = 0.5)
```

```python
sns.stripplot(data = penguins,x = 'species',y='body_mass_g',hue="island",dodge=True,jitter=True)
```

```python
# swarm plot
sns.set_context('talk')
sns.set_style("white")
sns.swarmplot(data = penguins,x = 'species',y='body_mass_g',hue="island")
sns.despine()
```

```python
# histogram
sns.set_context("notebook")
sns.histplot(data = penguins,x = 'body_mass_g',hue='sex',multiple = 'stack',palette='dark')

```

```python
# regression plot
sns.regplot(data = penguins,x= 'body_mass_g',y = 'flipper_length_mm',color='blue')
sns.set_palette('pastel')
```

```python
# line plot
sns.set_style('whitegrid')
sns.lineplot(data = penguins,x = 'body_mass_g',y = 'flipper_length_mm',hue = 'sex')
```

```python
# jointplot
sns.jointplot(data = penguins,x = 'body_mass_g',y='flipper_length_mm',color='blue')
```

```python
sns.jointplot(data = penguins,x = 'body_mass_g',y='flipper_length_mm',hue='sex',color='blue')
```

```python
sns.jointplot(data = penguins,x = 'body_mass_g',y='flipper_length_mm',hue='sex',color='blue',kind = 'hist')
```

```python
sns.jointplot(data = penguins,x = 'body_mass_g',y='flipper_length_mm',color='blue',kind = 'reg')
# kind decides what type of plot appears in the center.
```

```python
# bar plot
sns.barplot(data = penguins,x='species',y='body_mass_g',hue='sex',palette=['pink','green'],estimator=np.sum)
# hue is used to separate data into categories using colors.
```

```python
sns.barplot(data = penguins,x='species',y='body_mass_g',hue='sex',palette=['pink','green'],estimator=np.var)
# estimator parameter is used to decide how values are aggregated (summarized) when multiple observations exist for a category.
```

```python
# countplot
sns.countplot(data = penguins,x = 'species',hue='island',palette='Set2')
```

```python
# box plot
sns.boxenplot(data = penguins,x='species',y='body_mass_g',hue='island')
```

```python
# violin plot
sns.violinplot(data = penguins,x='species',y='body_mass_g',hue = 'sex',split =True) 
```

```python
sns.violinplot(data = penguins,x='species',y='body_mass_g') 
sns.swarmplot(data = penguins,x = 'species',y = 'body_mass_g',hue = 'island',palette='dark',size = 3)
# swarm plot over violin plot
```

```python
# kde (kernel desity estimation) plot 
# helps to understand where the data is concentrated and how it spreads
sns.kdeplot(data = penguins,x = 'body_mass_g',hue='species',palette='pastel',fill=True)
```

```python
columns = ['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g'] 	
penguins[columns].corr()
```

```python
# heat map: 2d color coded matrix that shows relationship or patterns b/w two or more numeric variables
# used to show correlation b/w the features
columns = ['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g'] 
sns.heatmap(data = penguins[columns].corr(),annot = True,cmap ='Blues',vmin = -0.3,linewidths=2,linecolor='black')
```

```python
# rug plot
sns.rugplot(data =penguins,x = 'body_mass_g',hue= 'species',palette='dark',height=0.05)
```

```python
sns.rugplot(data =penguins,y= 'body_mass_g',hue= 'species',palette='dark',height=0.5)
```

```python
# pair plot:shows all relationship b/w numeric variables
sns.pairplot(data = penguins)
```

```python
sns.pairplot(data = penguins,hue='sex',palette='Set2',diag_kind='kde')
```

```python
# pair grid->subplot grid for ploting  pairwise relationship in a dataset
graph = sns.PairGrid(data = penguins,hue = 'sex',palette='Set2')
graph.map_upper(sns.scatterplot)
graph.map_lower(sns.kdeplot)
graph.map_diag(sns.histplot)
graph.add_legend()

```
