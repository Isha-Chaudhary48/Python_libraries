```python
q1 = cleanData['stocks'].quantile(0.25)
q3 = cleanData['stocks'].quantile(0.75)
```

```python
q1
```

```python
q3
```

```python
IQR = q3 - q1
```

```python
IQR
```

```python
lower = q1 - 1.2 * IQR
upper = q3 + 1.2 * IQR
```

```python
lower
```

```python
upper
```

```python
cData = cleanData[(cleanData['stocks'] >=lower ) & (cleanData['stocks'] <= upper)]
```

```python
cData
```

```python
sns.boxplot(cData['stocks'])
plt.show()
```
