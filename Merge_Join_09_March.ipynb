```python
students = pd.DataFrame({
    'roll':[101,102,103,107,108],
    'name':['amit','sumit','pawan','ram','rajat']
})
marks=pd.DataFrame({
    'roll':[101,102,103,105,106,107,108,109],
    'marks':[45,65,78,45,32,98,56,68]
})
```

```python
pd.merge(students,marks,on="roll")
```

```python
pd.merge(students,marks,on="roll",how='right')
```

```python
pd.merge(students,marks,on="roll",how='outer')
```
